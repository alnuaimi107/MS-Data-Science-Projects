```python
import pandas as pd
import seaborn as sns
import numpy as np
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
```

```python
''' Load the data sets for:
1- Remote Automatic Weather Stations (RAWS) as df1
2- Wildland Fire Locations (WFL) as df2 
3- Texas Fire Department Substations'''

df1 = pd.read_csv("Public_View_-_Interagency_Remote_Automatic_Weather_Stations_(RAWS).csv")
df2 = pd.read_csv("WFIGS_-_Wildland_Fire_Locations_Full_History.csv")
df3 = pd.read_excel("TX_Fire_Departments_091522.xlsx", sheet_name = "TXFireDeptSubstations")
df4 = pd.read_excel("TX_Fire_Departments_091522.xlsx", sheet_name = "TX FireDepartments")




df1.head()

df2.head()

df3.head()

df4.head()
```

```python
# Exploratory Data Analysis: for Remote Automatic Weather Stations (RAWS)

df1.info()
df2.info()

# Select Specific columns

df1_2 = df1[['Latitude','Longitude','StationName','State','County',
             'RainAccumulation','WindSpeedMPH','WindDirDegrees',
             'AirTempStandPlace','RelativeHumidity'
             ,'SolarRadiation','StationID','ObservedDate']]

df2_1 = df2[['X','Y','FireDiscoveryDateTime','FireOutDateTime','FireCause',
             'POOState','POOCounty']]

df3_2 = df3[['FireDepartmentName','SubstationName','PhysicalAddress1','County','Latitude','Longitude']]

df4_1 = df4[['Name','PhysicalCity','PhysicalAddress1','County','Latitude','Longitude']]





df1_2.head()
df2_1.head()
df3_2.head()
df4_1.head()
```

```python
# Checking for dublicates or Null

df1_2.duplicated().sum()

df1_2.isnull().sum()

df2_1.duplicated().sum()

df2_1.isnull().sum()

df3_2.duplicated().sum()

df3_2.isnull().sum()
```

```python
# Dropping any row that has a missing values espicially in Location & County

df1_3 = df1_2.dropna(axis=0, how="any", thresh=None, subset=None, inplace=False)

df1_3.isnull().sum()

df2_2 = df2_1.dropna(axis=0, how="any", thresh=None, subset=None, inplace=False)

df2_2.isnull().sum()

# Dropping any row that has a duplicates

df2_3 = df2_2.drop_duplicates(subset=None, keep="first", inplace=False)

df2_3.duplicated().sum()


df3_2.rename(columns = {'SubstationName':'SubstationName/PhysicalCity'}, inplace = True)
df4_1.rename(columns = {'PhysicalCity':'SubstationName/PhysicalCity', 'Name':'FireDepartmentName'}, inplace = True)



# Concatenate the fire department and fire substations
df_concat = pd.concat([df3_2,df4_1], ignore_index=True)

df_concat
df_concat.duplicated().sum()
```

```python
# new data frame with split value columns
new = df1_3["ObservedDate"].str.split(" ", n = 1, expand = True)

new.head()
 
# Separate Date column from data frame
df1_3["Date"]= new[0]
 
# Separate Time column from data frame
df1_3["Time"]= new[1]
 
# Dropping old ObservedDate column
df1_3.drop(columns =["ObservedDate"], inplace = True)

df1_3.head()
```

```python
# new data frame with split value columns
new = df2_3["FireDiscoveryDateTime"].str.split(" ", n = 1, expand = True)
 
# Separate Date column from data frame
df2_3["FireDiscoveryDate"]= new[0]
 
# Separate Time column from data frame
df2_3["FireDiscoveryTime"]= new[1]
 
# Dropping old FireDiscoveryDateTime column
df2_3.drop(columns =["FireDiscoveryDateTime"], inplace = True)


# new data frame with split value columns
new = df2_3["FireOutDateTime"].str.split(" ", n = 1, expand = True)
 
# Separate Date column from data frame
df2_3["FireOutDate"]= new[0]
 
# Separate Time column from data frame
df2_3["FireOutTime"]= new[1]
 
# Dropping old FireOutDateTime column
df2_3.drop(columns =["FireOutDateTime"], inplace = True)

# new data frame with split value columns
new = df2_3["POOState"].str.split("-", n = 1, expand = True)
 
# Separate State column from data frame
df2_3["State"]= new[1]
 
# Dropping old POOState column
df2_3.drop(columns =["POOState"], inplace = True)

df2_3.head()
```

```python
# Rename data frame columns
df2_3.rename(columns = {'Y':'Latitude',
                        'X':'Longitude',
                       'POOCounty':'County'}, inplace = True)

df2_3.head()
```

```python
# new data frame with split value columns making data type integer to be analyzed
new = df1_3["RainAccumulation"].str.split(" ", n = 1, expand = True)
df1_3["RainAccumulation(Inches)"]= new[0]
df1_3.drop(columns =["RainAccumulation"], inplace = True)

new = df1_3["WindSpeedMPH"].str.split(" ", n = 1, expand = True)
df1_3["WindSpeed(MPH)"]= new[0]
df1_3.drop(columns =["WindSpeedMPH"], inplace = True)

new = df1_3["WindDirDegrees"].str.split(" ", n = 1, expand = True)
df1_3["WindDirection(Degrees)"]= new[0]
df1_3.drop(columns =["WindDirDegrees"], inplace = True)

new = df1_3["AirTempStandPlace"].str.split(" ", n = 1, expand = True)
df1_3["AirTempreture(Fahrenheit)"]= new[0]
df1_3.drop(columns =["AirTempStandPlace"], inplace = True)

new = df1_3["RelativeHumidity"].str.split(" ", n = 1, expand = True)
df1_3["RelativeHumidity(%)"]= new[0]
df1_3.drop(columns =["RelativeHumidity"], inplace = True)

new = df1_3["SolarRadiation"].str.split(" ", n = 1, expand = True)
df1_3["SolarRadiation(W/M^2)"]= new[0]
df1_3.drop(columns =["SolarRadiation"], inplace = True)

df1_3.head()
```

```python
# Changing data type

df1_3.info()

df1_4 = df1_3[df1_3["RainAccumulation(Inches)"].str.contains("NO")== False]
df1_5 = df1_4[df1_4["WindSpeed(MPH)"].str.contains("NO")== False]
df1_6 = df1_5[df1_5["WindDirection(Degrees)"].str.contains("NO")== False]
df1_7 = df1_6[df1_6["AirTempreture(Fahrenheit)"].str.contains("NO")== False]
df1_8 = df1_7[df1_7["RelativeHumidity(%)"].str.contains("NO")== False]
df1_9 = df1_8[df1_8["SolarRadiation(W/M^2)"].str.contains("NO")== False]



# Using dictionary to convert specific columns
convert_dict = {'RainAccumulation(Inches)': float,
                'WindSpeed(MPH)': float,
                'WindDirection(Degrees)': float,
                'AirTempreture(Fahrenheit)': float,
                'RelativeHumidity(%)': float,
                'SolarRadiation(W/M^2)': float
                
                }
 
df1_10 = df1_9.astype(convert_dict)

df1_10
```

```python
# Taking only Texas inputs or data

# Weather_Stations_Data
df1_11= df1_10[df1_10.State.str.startswith('TX')]

#Wildland_Fire_Locations
df2_4 = df2_3[df2_3.State.str.startswith('TX')]

```
